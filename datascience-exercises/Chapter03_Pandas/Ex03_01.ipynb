```python
# Creating list
name = ["A", "B", "C", "D", "E"]
marks = [52, 60, 65, 78, 85]
# Creating series from list
data = pd.Series(data = marks, index = name) # pd.Series(data, index) has this syntax
data
```

```python
# Creating dicktionary
grades = {"sem1": 3.15, "sem2": 3.35, "sem3": 3.75}
# Creating series from dicktionary
data = pd.Series(data = grades)
data
```
