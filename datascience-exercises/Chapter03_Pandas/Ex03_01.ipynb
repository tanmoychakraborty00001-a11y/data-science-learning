```python
# Creating dictionary
data = {"name": ["A", "B", "C", "D", "E"],
        "age": [30, 20, 25, 50, 52],
        "marks": [65, 70, 75, 88, 90],
        "city": ["Mumbai", "Pune", "Chennai", "Delhi", "London"]}
# Creating dataframe from dicktionary
df = pd. DataFrame(data = data)
df 
```
