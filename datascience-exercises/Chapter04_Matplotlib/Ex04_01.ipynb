```python
# creating data points along axis
x = [2, 4, 6, 8] # data points along x axis
y = [3, 6, 9, 12] # data points along y axis
plt.plot(x, y)
```

```python
x = [5, 2, 3, 2, 4]
y = [1, 2, 3, 4, 5]
plt.figure(figsize=(10, 15)) # size of the figure using figure method 
# and passing values using tuples for making length of rows and columns
plt.plot(x, y)
plt.title("Data Info")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

## **Bar graphs**

```python
# Data
grades = ["A", "B", "C", "D", "E"]
values = [90, 75, 55, 50, 45]

# plot
plt.bar(grades, values, color="orange", edgecolor="black", alpha=0.8, width=0.6)
plt.title("Students grade reports")
plt.xlabel("Student grades")
plt.ylabel("Marks")
plt.show()


```
