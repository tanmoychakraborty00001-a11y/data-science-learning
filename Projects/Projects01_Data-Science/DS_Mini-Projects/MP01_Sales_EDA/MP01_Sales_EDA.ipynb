```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df['Order Date']
```

```python
df['Order Date'].dtype
```

```python
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df['Ship Date']
```

```python
df['Ship Date'].dtype
```

```python
df['Order Year'] = df['Order Date'].dt.year
df['Order Year']
```

```python
df['Order Month'] = df['Order Date'].dt.month
df['Order Month']
```

```python
df['Delivery Days'] = (df['Ship Date'] - df['Order Date']).dt.days
df['Delivery Days']
```

```python
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()
monthly_sales
```

```python
monthly_sales.plot(figsize=(12,5), title = "Monthly Sales Trend")
```

```python
monthly_sales.plot(figsize=(12,5), kind='bar', title = 'Monthly Sales Trend')
```

```python
month_sales = df.groupby(df['Order Date'].dt.month)['Sales'].sum()
month_sales
```

```python
month_sales.plot(figsize=(12,5), title='Monthwise Sales')
```

```python
month_sales.plot(figsize=(12,5), kind='bar', title='Monthwise Sales')
```

```python
yearly_sales = df.groupby(df['Order Date'].dt.year)['Sales'].sum()
yearly_sales
```

```python
yearly_sales.plot(figsize=(12,5), title='Yearly Sales Trend')
```

```python
yearly_sales.plot(figsize=(12,5), kind='bar', title='Yearly Sales Bar Plot')
```

```python
df['Delivery Days'].describe()
```

```python
df.groupby('Ship Mode')['Delivery Days'].mean().plot(figsize=(12,5), kind='bar', title='Average Delivery Days by Ship Mode')
plt.ylabel('Days')
plt.show()
```

```python
category_summary = df.groupby('Category')[['Sales', 'Profit']].sum()
category_summary
```

```python
category_summary[['Sales', 'Profit']].plot(figsize=(12,5), kind='bar', title='Sales by Category')
```

```python
df.groupby('Region')['Sales'].sum().plot(figsize=(12,5), kind='bar', title='Region-wise Sales')
plt.ylabel('Sales')
plt.show()
```

```python
top10_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
top10_products
```

```python
top10_products.plot(figsize=(12,5), kind='barh', title='Top 10 Products by Sales')
plt.xlabel('Sales')
plt.show()
```

```python
plt.scatter(df['Sales'], df['Profit'])
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Profit vs Sales')
plt.show()
```

```python
segment_summary = df.groupby('Segment')[['Sales', 'Profit']].sum()
segment_summary
```

```python
segment_summary.plot(figsize=(12,5), kind='bar', title='Sales by Customer Segment')
plt.show()
```
