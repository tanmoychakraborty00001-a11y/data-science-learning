```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df['Order Date']
```

```python
df['Order Date'].dtype
```

```python
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df['Ship Date']
```

```python
df['Ship Date'].dtype
```

```python
df['Order Year'] = df['Order Date'].dt.year
df['Order Year']
```

```python
df['Order Month'] = df['Order Date'].dt.month
df['Order Month']
```

```python
df['Delivery Days'] = (df['Ship Date'] - df['Order Date']).dt.days
df['Delivery Days']
```

```python
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()
monthly_sales
```

```python
monthly_sales.plot(figsize=(12,5), title = "Monthly Sales Trend")
```

```python
monthly_sales.plot(figsize=(12,5), kind='bar', title = 'Monthly Sales Trend')
```

```python
month_sales = df.groupby(df['Order Date'].dt.month)['Sales'].sum()
month_sales
```

```python
month_sales.plot(figsize=(12,5), title='Monthwise Sales')
```

```python
month_sales.plot(figsize=(12,5), kind='bar', title='Monthwise Sales')
```

```python
yearly_sales = df.groupby(df['Order Date'].dt.year)['Sales'].sum()
yearly_sales
```

```python
yearly_sales.plot(figsize=(12,5), title='Yearly Sales Trend')
```

```python
yearly_sales.plot(figsize=(12,5), kind='bar', title='Yearly Sales Bar Plot')
```
